## ARIMA Model


Time series forecasting using models like ARIMA can help predict future values based on historical data, which in this case is about AI trends such as AI software revenue, global market value, adoption rates, and more.

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Fitting ARIMA model for AI Software Revenue prediction
model = ARIMA(df['AI Software Revenue(in Billions)'], order=(1, 1, 1))
arima_result = model.fit()

# Forecast future values
forecast = arima_result.forecast(steps=3)  # Predict the next 3 years
print(forecast)

# Plot forecast
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['AI Software Revenue(in Billions)'], label='Actual')
plt.plot([2026, 2027, 2028], forecast, label='Forecast', linestyle='--')
plt.title('AI Software Revenue Forecast')
plt.xlabel('Year')
plt.ylabel('AI Software Revenue (in Billions)')
plt.legend()
plt.show()
```

By using time series forecasting (ARIMA), you can predict how certain AI trends will evolve in the future. This can be particularly useful for planning business strategies, understanding market growth, or even estimating the impact of AI on employment and revenue in the coming years.

### Split Y and X values 

```python
# Features: Year (independent variable)
X = df[['Year']]

# Target: AI Software Revenue (dependent variable)
y = df['AI Software Revenue(in Billions)']
```

```python
from sklearn.linear_model import LinearRegression

# Create the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Print the slope (coefficient) and intercept
print(f"Slope (Coefficient): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
# Predict future AI Software Revenue for the next few years (e.g., 2026-2030)
future_years = pd.DataFrame({
    'Year': [2026, 2027, 2028, 2029, 2030]
})

# Make predictions using the model
predictions = model.predict(future_years)

# Print the predictions
for year, prediction in zip(future_years['Year'], predictions):
    print(f"Predicted AI Software Revenue for {year}: ${prediction:.2f} billion")
```

```python
import matplotlib.pyplot as plt

# Plot the original data
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Plot the future predictions
plt.scatter(future_years, predictions, color='green', label='Future Predictions')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('AI Software Revenue (Billions)')
plt.title('AI Software Revenue Prediction using Linear Regression')
plt.legend()

# Show the plot
plt.show()
```

## **Evaluation and Validation**
<a href=#cont>Back to Table of Contents</a>

* **Purpose:** Evaluate and validate the effectiveness and accuracy of the models.
* **Details:** Present metrics used to evaluate the models, such as accuracy, precision, recall, F1-score, etc. Discuss validation techniques employed, such as cross-validation or train/test split.
---

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Predictions for the existing data
y_pred = model.predict(X)

# R-squared score
r2 = r2_score(y, y_pred)
print(f'R-squared: {r2:.2f}')

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error (MSE): {mse:.2f}')

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
```
